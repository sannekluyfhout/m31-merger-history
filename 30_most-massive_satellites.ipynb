```python
#again i want to sort by peak mass, descending, among destroyed satellites, so i can find the most massive ones that were destroyed
peak_mass_values = sat_peak_mass.values  # shape (302,)
peak_mass_masked = np.where(destroyed_mask, peak_mass_values, -np.inf)
top_n = 30 #printing top 30 to see if it makes sense
top_indices = np.argsort(peak_mass_masked)[::-1][:top_n]
```

```python
def sf_to_lookback(a):
    """Convert scale factor(s) to lookback time in Gyr."""
    a = np.asarray(a, dtype=float)
    z = 1.0 / a - 1.0
    return Planck18.lookback_time(z).to_value(u.Gyr)
```

```python
print(f"Selected top {top_n} most massive destroyed satellites:")
for rank, idx in enumerate(top_indices):
    m_peak = peak_mass_values[idx] * 1e10   # in Msun
    a_dest = float(mergers_ds.sel(
        snap_subhalo_id=idx, which='subhalo', event='destroyed'
    ).scale_factor.values)
    t_dest = sf_to_lookback(a_dest)
    print(f"  #{rank+1:2d}  index={idx:3d}  peak_mass={m_peak:.2e} Msun  "
          f"destroyed at lookback {t_dest:.2f} Gyr ago")
```

I also believe this makes sense and that subhalo id 50 the subhalo of m31 is. We are still looking at the point where the sattelites have the highest mass. 
I am accounting for change of id number by using "mergers_m31.nc" this is indexed by snap_subhalo_id, and is tracking 1 merger event tracked in time. The subhalo id stored at each event is the id that is for that specific snapshot, but since i do it for r200_first I get the mass of the subhalo at that point.

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.cosmology import Planck18
from astropy import units as u

def sf_to_lookback(a):
    a = np.asarray(a, dtype=float)
    z = 1.0 / a - 1.0
    return Planck18.lookback_time(z).to_value(u.Gyr)

#choose event where to define the mass ratio
event = "sat_highestmass"   
top_n = 30

#masses at peak satellite mass
m_sat = ds["SubhaloMass"].sel(event=event, which="subhalo").values * 1e10 / mt.h
m_main = ds["SubhaloMass"].sel(event=event, which="main").values * 1e10 / mt.h
mu = m_sat / m_main

#time of destruction
a_dest = ds["scale_factor"].sel(event="destroyed", which="subhalo").values
t_dest = sf_to_lookback(a_dest)

#only destroyed satellites
destroyed = ~ds["survives_to_now"].values

#select top 30 most massive destroyed satellites
mask = destroyed & np.isfinite(m_sat) & np.isfinite(mu) & (m_sat > 0)
order = np.argsort(np.where(mask, m_sat, -np.inf))[::-1]
top_indices = order[:top_n]

#values for plotting
top_mu = mu[top_indices]
top_time = t_dest[top_indices]
top_mass = m_sat[top_indices]
ranks = np.arange(1, top_n + 1)

fig, ax = plt.subplots(figsize=(9, 6))

#mass-ratio ranges
ax.axhspan(0.05, 0.10,
           color="gold", alpha=0.18,
           label=r"$0.05\leq\mu<0.10$")

ax.axhspan(0.10, 0.20,
           color="orange", alpha=0.18,
           label=r"$0.10\leq\mu<0.20$")

ax.axhspan(0.20, 0.50,
           color="red", alpha=0.12,
           label=r"$0.20\leq\mu\leq0.50$")

sc = ax.scatter(
    ranks,
    top_mu,
    c=top_time,
    s=80,
    edgecolor="black",
    zorder=5
)

ax.axhline(0.05, linestyle="--", linewidth=1)
ax.axhline(0.50, linestyle="--", linewidth=1)

ax.set_yscale("log")
ax.set_xlabel("Rank by peak satellite mass")
ax.set_ylabel(r"Mass ratio $\mu = M_{\rm r200}/M_{\rm main}$")
ax.set_title(f"Mass ratios of top {top_n} most massive destroyed mergers")
ax.grid(True, alpha=0.3)

cbar = plt.colorbar(sc, ax=ax)
cbar.set_label("Destruction lookback time [Gyr]")

ax.legend(loc="upper right", fontsize=8)
plt.tight_layout()
plt.show()
```
